```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# A simple scatter plot with axes labeled and title

x = [1, 2, 3, 4, 5]
y = [5, 6, 7, 8, 9]

plt.xlabel('x-axis title')  # Labels must be defined before the .show() method
plt.ylabel('y-axis title')
plt.title('My plot title')
plt.scatter(x, y)  # x and y lists must be the same length.
plt.show()
```

```python
# Two sets of data with legend

x = [1, 2, 3, 4, 5]
y1 = [5, 6, 7, 8, 9]
y2 = [7, 2, 5, 3, 8]

plt.xlabel('x-axis title')
plt.ylabel('y-axis title')
plt.title('My plot title')
plt.scatter(x, y1, label = 'data 1')  # Must include .legend() method for the labels to show.
plt.scatter(x, y2, label = 'data 2')
plt.legend()
#plt.legend(loc=(1.04,0.93))  # Specify legend location outside of the canvas
plt.show()
```

```python
# Gridline options

x = [1, 2, 3, 4, 5]
y = [5, 6, 7, 8, 9]

plt.scatter(x, y)
#plt.grid()
#plt.grid(axis = 'x')  # Gridlines in one direction only. Option: 'y'
plt.grid(linestyle = '-.', alpha = 0.6, linewidth = 0.7)  # Dashed lines, faded, and smaller width
plt.show()
```

```python
# Polynomial fit with numpy
np.random.seed(12)

x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [5, 3, 7, 2, 12, 6, 14, 4, 8]
p = np.poly1d(np.polyfit(x, y, 2))  # Fit data to N-degree polynomial (1 = linear)

mn = np.min(x)
mx = np.max(x)

t = np.linspace(mn, mx, 200)
plt.plot(x, y, 'o', t, p(t), '--')
plt.show()
```

```python
# Setting canvas features (size, labels, font, etc)

from matplotlib.pyplot import figure

x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [5, 3, 7, 2, 12, 6, 14, 4, 8]
p = np.poly1d(np.polyfit(x, y, 2))  # Fit data to N-degree polynomial (1 = linear)

mn = np.min(x)
mx = np.max(x)

t = np.linspace(mn, mx, 200)

# Figsize(x, y) in units of inches. dpi = 'dots per inch'
figure(figsize=(10, 6), dpi = 80)  # This must come before labels.

plt.tick_params(labelsize = 14)
plt.rcParams["font.family"] = "Arial"

plt.xlabel('x-axis title (units)', fontsize = 16, loc = 'right', labelpad = 12)  # Size of labels and title can be adjusted.
plt.ylabel('y-axis title (units)', fontsize = 16, loc = 'top', labelpad = 12)    # Title position can be adjusted.
plt.title('My plot title', fontsize = 20, loc = 'right', pad = 12)  # Pad and labelpad set distance between tick and axes labels.

plt.scatter(x, y) 
plt.plot(x, y, 'o', t, p(t), '--')
plt.savefig('test.png', dpi = 100)  # Savefig must be before .show(), or saved filed will be blank.
plt.savefig('test.pdf', dpi = 100)

plt.show()
```

```python
# Two vertically stacked plots using plt.subplot()

figure(figsize=(11, 7), dpi = 80)

x = np.asarray(x) ; y = np.asarray(y)  # (x, y) must be numpy arrays, not lists

plt.subplot(2, 1, 1)
plt.plot(x, y, 'o-')
plt.title('Title', fontsize = 20, loc = 'right', pad = 12)
plt.ylabel('y-label1', fontsize = 16, loc = 'top', labelpad = 12)

plt.subplot(2, 1, 2)
plt.plot(x, -y, 'o-')
plt.xlabel('x-label', fontsize = 16, loc = 'right', labelpad = 12)
plt.ylabel('y-label2', fontsize = 16, loc = 'top', labelpad = 12)


plt.show()
```

```python
# Two horizontally stacked plots using plt.subplot()

figure(figsize=(11, 7), dpi = 80)

x = np.asarray(x) ; y = np.asarray(y)  # (x, y) must be numpy arrays, not lists

plt.subplot(1, 2, 1)
plt.plot(x, y, 'o-')
plt.ylabel('y-label', fontsize = 16, loc = 'top', labelpad = 12)
plt.title('Title1', fontsize = 20, loc = 'right', pad = 12)

plt.subplot(1, 2, 2)
plt.plot(x, -y, 'o-')
plt.xlabel('x-label', fontsize = 16, loc = 'right', labelpad = 12)
plt.title('Title2', fontsize = 20, loc = 'right', pad = 12)

plt.show()
```

```python
# Four subplots using plt.subplots()

fig, axs = plt.subplots(2, 2, figsize=(12,10))
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Title 1', fontsize = 20)
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Title 2', fontsize = 20)
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Title 3', fontsize = 20)
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Title 4', fontsize = 20)

for index, ax in enumerate(axs.flat):        
    ax.set(xlabel='x-label', ylabel='y-label')
    
    for index, item in enumerate(([ax.xaxis.label, ax.yaxis.label])):
        item.set_fontsize(16)
    
# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```
