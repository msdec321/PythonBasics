```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in the data
filename = 'owid-covid-data.csv'
path = 'C:/Users/Matth/git/DataAnalysisWorkbooks/Covid19/Data/Raw_data/'

data = pd.read_csv(path+filename) 
```

```python
# Make a scatter plot of the data (A simple scatter plot can be messy if the data is unordered)

data.plot(x = 'date', y = 'total_cases', style = '.', figsize = (15,5))
```

```python
# The combination of "set_index" and the .loc() method can be used to apply selections to the data

df1 = data.set_index('location')
df1.loc['World'].plot(x = 'total_cases', y = 'total_deaths', style = '.', figsize = (15,5))
df1.loc['World'].plot(x = 'date', y = 'total_cases', style = '.', figsize = (15,5))
```

```python
# Using multiple selections (Note: The data are combined and plotted as a single time series)

_ = df1.loc['Africa' and 'Asia'].plot(x = 'date', y = 'new_cases', style = '.', figsize = (15,5), title = '') 
```

```python
# Multiple data series on a single canvas

ax = df1.loc['Africa'].plot(x = 'date', y = 'new_cases', style = '.', figsize = (15,5), label = 'Africa', title = '') 
df1.loc['Asia'].plot(x = 'date', y = 'new_cases', style = '.', figsize = (15,5), label = 'Asia', ax=ax) 
df1.loc['Europe'].plot(x = 'date', y = 'new_cases', style = '.', figsize = (15,5), label = 'Europe', ax=ax) 
```

```python
# Dividing two columns and plotting the result
data["cases_rate"] = data['total_cases'] / data['population']  # Add new column to dataframe
df1 = data.set_index('location')

_ = df1.loc['United States'].plot(x = 'date', y = 'cases_rate', style = '.', figsize = (15,5), title = '') 


data["caseDeathRatio"] = data['total_deaths'] / data['total_cases']  # Add new column to dataframe
df1 = data.set_index('location')

ax = df1.loc['North America'].plot(x = 'date', y = 'caseDeathRatio', style = '.', figsize = (15,5), label = 'North America') 
df1.loc['South America'].plot(x = 'date', y = 'caseDeathRatio', style = '.', figsize = (15,5), label = 'South America', ax=ax) 

# Optionally, adjust the y-axis range
ax.set_ylim(0.01, 0.09)
```
